```python
import pandas as pd
```

```python
# 1. Crie uma tabela contendo as seguintes informações:
# nome	departamento	idade	salario
# Ana	Vendas	28	3500
# Carlos TI	34	5500
# Beatriz	Marketing	26	3200
# Daniel	Recursos Humanos	45	6200
# Eduardo	TI	29	4800

tabela = {
    "nome": ["Ana", "Carlos", "Beatriz", "Daniel", "Eduardo"],
    "departamento": ["vendas", "TI", "marketing", "RH", "TI"],
    "idade": [28, 34, 26, 45, 29],
    "salario": [3500, 5500, 3200, 6200, 4800]
}
df = pd.DataFrame(data=tabela)
df
```

```python
# 2. Adicione a seguinte coluna à tabela:
# anos_de_experiencia
# 5 / 10 / 3 / 20 / 7
df["anos_de_experiencia"] = [5,10,3,20,7]
df
```

```python
# 3. Imprima somente o nome e salário dos funcionários
# maiores de 30 anos.

maiores = df[df["idade"] > 30] #filtro
maiores[["nome", "salario"]]
```

```python
# 4. Qual a média, mediana, moda e desvio padrão
# dos salários dos funcionários?

import statistics as stat

df_salario = df["salario"]
media = stat.mean(df_salario)
moda = stat.mode(df_salario)
mediana = stat.median(df_salario)
desvio = stat.stdev(df_salario)

print(f"media: {media}")
print(f"moda: {moda}")
print(f"mediana: {mediana}")
print(f"desvio: {round(desvio,2)}")
```

```python
# 5. Qual funcionário recebe o maior salário?
# Ele é o funcionário mais velho?

maiorSalario = max(df["salario"])
maiorIdade = max(df["idade"])

dadosFuncionario = df[df["salario"] == maiorSalario] #filtro
nome = dadosFuncionario["nome"].values[0]
idade = dadosFuncionario["idade"].values[0]

print(f"{nome} é o funcionario que tem maior salario")

if idade == maiorIdade:
  print("Sim, o funcionario mais velho recebe mais")
else:
  print("Nao, ele nao é o mais velho")
```

```python
import pandas as pd
import io
import requests
url = "https://raw.githubusercontent.com/robertotomchak/imigrantesavancado/main/dados_enem.csv"
s = requests.get(url).content
df2 = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

Utilizando os dados do ENEM responda:

```python
# 6. Quantos candidados de Curitiba fizeram a prova?
df_curitiba = df2[df2["NO_MUNICIPIO_ESC"] == "Curitiba"]
print(len(df_curitiba))
```

```python
# 7. Quantos estudantes não fizeram a prova? Isto é, zeraram todas as provas.
df2["NOTA_MEDIA"] = (df2["NU_NOTA_CH"] + df2["NU_NOTA_CN"] + df2["NU_NOTA_LC"] + df2["NU_NOTA_MT"] + df2["NU_NOTA_REDACAO"]) / 5
df_zerou = df2[df2["NOTA_MEDIA"] == 0]
print(len(df_zerou))
```

```python
# 8. Dado um estudante aleatório, qual a probabilidade dele ter uma nota maior que 700 em Matemática?
df_maior700 = df2[df2["NU_NOTA_MT"] > 700]
probabilidade = len(df_maior700) / len(df2) * 100
print(f"{round(probabilidade, 2)}%")
```

```python
# 9. Compare a porcentagem de estudantes que escolheram inglês e espanhol como língua estrangeira.
# Qual a língua estrangeira mais escolhida?
df_ingles = df2[df2["TP_LINGUA"] == "Ingles"]
df_espanhol = df2[df2["TP_LINGUA"] == "Espanhol"]

print(f"Ingles: {len(df_ingles)}")
print(f"Espanhol: {len(df_espanhol)}")

if len(df_ingles) > len(df_espanhol):
  print("Ingles foi a lingua mais escolhida")
else:
  print("Espanhol foi a lingua mais escolhida")
```

```python
# 10. Quantos estudantes tiveram nota na redação igual a 1000?
# A maioria era de escola pública ou privada?
df_redacao1000 = df2[df2["NU_NOTA_REDACAO"] == 1000]
df_redacao1000_publica = df_redacao1000[df_redacao1000["TP_ESCOLA"] == "Publica"]
df_redacao1000_privada = df_redacao1000[df_redacao1000["TP_ESCOLA"] == "Privada"]

print(f"Redações Nota 1000: {len(df_redacao1000)}")
print(f"De escola pública: {len(df_redacao1000_publica)}")
print(f"De escola privada: {len(df_redacao1000_privada)}")
if len(df_redacao1000_publica) > len(df_redacao1000_privada):
  print("Maioria foi de escola pública")
else:
  print("Maioria foi de escola privada")
```
